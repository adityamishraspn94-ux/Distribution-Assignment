###**Q1. Simulate 30 rolls with =RANDBETWEEN(1,6). What is the probability of rolling a 3 exactly 5 times? (Use BINOM.DIST)**

###**Answer:**
To find the probability of getting **3 exactly 5 times in 30 rolls**, use the formula:

```
=BINOM.DIST(5, 30, 1/6, FALSE)
```

This gives the probability of getting exactly **5 threes** out of **30** rolls.

###**Q4. What is the concept of standardization (z-score), and why is it important in data analysis? Explain the formula and how standardization transforms a dataset.**

###**Answer:**
Standardization (z-score) is a method used to **convert data into a common scale** without changing the shape of the distribution. It tells us **how many standard deviations** a value is away from the mean.

The formula is:

[
z = \frac{x - \mu}{\sigma}
]

Where:

* **x** = individual value
* **μ (mu)** = mean of the dataset
* **σ (sigma)** = standard deviation

**Why it is important:**

* It helps compare values from different scales.
* It is used in many algorithms like regression, clustering, and machine learning.
* It makes the data easier to analyze, especially when variables have different units.

**How standardization transforms a dataset:**

* The **mean becomes 0**.
* The **standard deviation becomes 1**.
* The new dataset is in **z-scores**, showing how far each value is from the mean in standard deviation units.

This makes data consistent and easier to interpret across different measurements.

###**Q5. What is Kurtosis and their types?**

###**Answer:**
Kurtosis is a statistical measure that describes the **shape of a distribution’s tails**—that is, how **heavy or light** the tails are compared to a normal distribution. It tells us whether the data have more outliers or fewer outliers.

###**Types of Kurtosis:**

1. **Mesokurtic (K = 0)**

   * Shape similar to the normal distribution
   * Moderate tails and moderate peak

2. **Leptokurtic (K > 0)**

   * Distribution has **heavy tails**
   * More outliers and a **sharper peak**
   * Data is more concentrated in the center but has extreme values in tails

3. **Platykurtic (K < 0)**

   * **Light tails** and **flatter peak**
   * Fewer outliers
   * Data is more evenly spread out

In simple terms, kurtosis helps us understand **how extreme the extremes are** in a dataset.

###**Q6. Explain why the uniform distribution is a good model for the outcome of rolling a fair die.**

###**Answer:**
A uniform distribution is a good model for a fair die because **each outcome has an equal probability of occurring**.
For a fair six-sided die, the numbers 1 to 6 all have the same chance:

[
P(1)=P(2)=P(3)=P(4)=P(5)=P(6)=\frac{1}{6}
]

This matches the definition of a **uniform distribution**, where all possible values are **equally likely**.
Therefore, the roll of a fair die follows a **discrete uniform distribution**.

###**Q7. Use Excel to compute the probability of getting at least 8 successes in 15 trials with success probability 0.5.**

###**Answer:**
To calculate the probability of getting **at least 8 successes** in 15 trials when **p = 0.5**, use the **BINOM.DIST** or **BINOM.DIST.RANGE** function.

The easiest method is:

```
=BINOM.DIST.RANGE(15, 0.5, 8, 15)
```

This gives the probability of getting **8 to 15 successes** out of 15 trials.

If your Excel version doesn't support BINOM.DIST.RANGE, use:

```
=1 - BINOM.DIST(7, 15, 0.5, TRUE)
```

This subtracts the cumulative probability of 0–7 successes from 1.

This result gives the probability of **at least 8 successes**.

###**Q8. How does log transformation help in stabilizing variance and making data more normally distributed?**

###**Answer:**
A **log transformation** is often used in data analysis to handle skewed data and reduce the impact of large values.

###**How it helps:**

1. **Stabilizes Variance:**
   In many datasets, larger values have higher variability.
   Taking the log compresses these large values, reducing their spread and making the variance more stable across the dataset.

2. **Reduces Right Skewness:**
   Data that are heavily right-skewed (with a long tail of large values) become more balanced after log transformation.
   This helps the data look more **symmetrical** and closer to a **normal distribution**.

3. **Handles Outliers Better:**
   Extreme high values become less extreme after taking logs, making the dataset easier to analyze.

**In summary:**
Log transformation **compresses large values, reduces skewness, stabilizes variance**, and often makes the data **more normally distributed**, which is useful for statistical modeling and machine learning.
